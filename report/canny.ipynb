# Canny detector for fused_tp0_ch0+ch1_v2

```python
import numpy as np
import pandas as pd
from utils import load_dataset, aggregate, plot_quantiles, plot_surface, plot_common
import matplotlib.pyplot as plt
from IPython.display import Markdown
from IPython.display import display, Markdown

algorithm = 'canny'
legend = "2048x2048x{n}"
method = 'mpisingle'
pd.set_option('display.max_rows', 10000)
data = load_dataset('csv', 'canny_fused_tp0')

def aggregate(data, stat, aggfunc):
    maxes = data[(data['stat'] == stat) & (data['method'] == method)].pivot_table('time_ms', columns=['size'], index=['round'], aggfunc='max')
    return maxes.reset_index().melt(id_vars=['round'], value_name='time_ms').pivot_table('time_ms', columns=['size'], aggfunc=aggfunc)
```

## reading the dataset

```python
read_ds = (data[(data['stat'] == 'read') & (data['method'] == 'mpisingle')]['time_ms'] / 1000)
plt.boxplot(read_ds)
plt.xticks([1], ["read"])
plt.ylabel('time [s]')
plt.savefig(f"figures/canny/read_boxplot.pdf")
read_ds.describe()
read_ds.quantile([0.05, 0.50, 0.95])
```

```python
for stat in data['stat'].unique():
    display(Markdown("# " + stat))
    p2, p5, p8 = [(aggregate(data, stat, lambda x: x.quantile(q))/1000).T for q in [0.2, 0.5, 0.8]]

    for c in [p2, p5, p8]:
        table = pd.DataFrame({
            'time_s': c['time_ms'],
            'speedup': (c.T[1] / c)['time_ms']
        })
        table.index.rename('ranks', inplace=True)
        display(table)
        print(table.to_latex())
    
    plt.plot(p5, marker='o')
    plt.fill_between(p5.index, p8['time_ms'], p2['time_ms'], alpha=0.5)
    plt.ylabel('time [s]')
    plt.xlabel('ranks')
    plt.savefig(f"figures/canny/time_{stat}.pdf")
    plt.show()
    
    (p5.T[1] / p5).plot(marker='o')
    plt.plot([1, 2, 4, 6, 8], [1, 2, 4, 6, 8], marker='o')
    plt.ylabel('speedup')
    plt.xlabel('ranks')
    plt.legend(labels=['MPI enabled Op', 'ideal'])
    plt.savefig(f"figures/canny/speedup_{stat}.pdf")
    plt.show()
```
